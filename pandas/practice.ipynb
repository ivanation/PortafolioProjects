# Group by and Aggregating

```python
da = pd.read_csv('Flavors.csv')
da
```

```python
group_by_frame = da.groupby('Base Flavor')
group_by_frame.mean()
```

```python
da.groupby('Base Flavor').count()
```

```python
da.groupby('Base Flavor').min()
```

```python
da.groupby('Base Flavor').max()
```

```python
da.groupby('Base Flavor').sum()
```

```python
da.groupby('Base Flavor').agg({'Flavor Rating':['mean','max','count','sum']})
```

```python
 da.groupby('Base Flavor').agg({'Flavor Rating':['mean','max','count','sum'], 'Texture Rating':['mean','max','count','sum']})
```

```python
da
```

```python
da.groupby(['Base Flavor', 'Liked'])
```

```python
da.groupby(['Base Flavor', 'Liked']).mean()
```

```python
da.groupby(['Base Flavor', 'Liked']).agg({'Flavor Rating':['mean','max','count','sum']})
```

```python
da.groupby('Base Flavor').describe()
```
